```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

# Cargar el conjunto de datos limpio
data = pd.read_csv('Cleaned-Data.csv')

# Separar las características (X) y la variable objetivo (y)
X = data.drop(columns=['Severity_Mild', 'Country'])  # Características
y = data['Severity_Mild']  # Variable objetivo

clf = DecisionTreeClassifier(random_state=42)

# Entrenar el modelo
clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree


plt.figure(figsize=(15, 10))
plot_tree(clf, filled=True, feature_names=X.columns.tolist(), class_names=['Not Mild', 'Mild'])
plt.show()
```
